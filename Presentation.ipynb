# My plan



<b> Data:</b> European Survey Study 2016<br>

<b> Objective:</b> Build four different machine learning models to predict participants beliefs on climate change<br>



# About the European Social Survey

<ul>
    <li> Multi-country survey that aims to unravel public opinions wide range of topics, like:</li>
    <ul>
        <li> Attitudes and values towards media, social trust and politics</li>
        <li> 2016 edition: attitude towards climate change </li>
    </ul>
    <li> Participants education, job-status, gender, age etc. are recorded</li>
    <li> ~370 features</li>
    <li> >44.000 participants</li>

 Choose four supervised Classifcation Models
NO Logistic Regression: multicollinearity is inescapable with so many variables
<ul>
    <li> <b>Decision Trees Classifier</b> </li>
    <ul>
        <li> Flexible </li>
        <li> Resilient to noise (irrelevant features) </li>
    </ul>
    <li> <b>Random Forest Classifier</b> </li>
    <ul>
        <li> Flexible and fast </li>
        <li> More resilient to multicollinearity than trees </li>
        <li> Can handle imbalanced datasets </li>
        <li> Option to extract feature importance </li>
    </ul>
    <li><b> Support vector Classifier</b></li>
    <ul>
        <li> Takes too much time to compute! Up to 2 hours per round </li>
    </ul>
    <li> <b>XGBoost</b> </li>
    <ul>
        <li> Also a type of tree</li>
        <li> Very fast </li>
        <li> Difficult to get the parameters just right, but is said to win most Kaggle competitions </li>
    </ul>
